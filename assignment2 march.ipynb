Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib

AnsMatplotlib is a Python data visualization library used to create static, animated, and interactive visualizations in Python. It is one of the most widely used plotting libraries in the scientific and data science communities. It can be used to create a wide range of plots, including line plots, scatter plots, bar plots, histograms, and more.

Matplotlib's pyplot module is a collection of functions that allow you to create a variety of plots quickly and easily. Some of the plots that can be plotted using the Pyplot module of Matplotlib are:

Line plot - A basic plot where data is plotted as a series of points connected by lines.
Scatter plot - A plot where individual data points are plotted as markers on a coordinate grid.
Bar plot - A plot where data is represented using rectangular bars.
Histogram - A plot where data is divided into intervals, and the number of values falling within each interval is plotted as a bar.
Pie chart - A plot where data is represented as a circle, with each section of the circle representing a different category of data and its size corresponding to the proportion of the data in that category.
Overall, Matplotlib and its Pyplot module provide a powerful toolset for data visualization in Python, enabling users to create professional-quality plots and visualizations to help communicate their data effectively.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

Ans A scatter plot is a type of plot used to display the relationship between two variables. It consists of a set of points, where each point represents an observation in the data and its location on the plot corresponds to the values of the two variables being plotted.

To generate a scatter plot using the given code in Python, we can use the Matplotlib library's Pyplot module. Here's an example code to generate a scatter plot using the given data:

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(3)
x = 3 + np.random.normal(0,2,50)
y = 3 + np.random.normal(0,2,len(x))
plt.scatter(x,y)
plt.title(" scatter plot" )
plt.xlabel(" x axis")
plt.ylabel(" y axis")
plt.show()
```

This will produce a scatter plot with x-axis, y-axis labels, and a title. The resulting plot will show the relationship between x and y for the generated data. The scatter plot will have dots or markers corresponding to each value of x and y. The distribution and correlation between the two variables can be visually analyzed from this plot

Here's an example code to draw a bar plot and a horizontal bar plot using the following data:

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.barh(company,profit)
plt.title("company profit")
plt.xlabel("profit")
plt.ylabel("company")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

Ans A box plot is a type of graphical representation that displays the distribution of data based on five summary statistics: minimum value, first quartile (25th percentile), median (50th percentile), third quartile (75th percentile), and maximum value. The box represents the interquartile range (IQR), which is the difference between the third and first quartiles. The whiskers extending from the box represent the minimum and maximum values that fall within 1.5 times the IQR. Any data point that falls outside of the whiskers is considered an outlier and plotted as a point.

Box plots are commonly used in data analysis and statistical inference to identify the central tendency, variability, and outliers of a data set. They can be used to compare the distribution of data between different groups or to visualize the distribution of a single variable.

To plot a box plot of the given data, we first need to import the necessary libraries:

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1,box2])
plt.title("Box plot of two data sets")
plt.xticks([1,2],["Box 1", "Box 2"])
plt.ylabel("Values")
plt.show()
```

This will create a box plot with two boxes, one for each data set, labeled "Box 1" and "Box 2" on the x-axis. The y-axis represents the values in the data sets. The box represents the IQR, the whiskers extend to the minimum and maximum values within 1.5 times the IQR, and any outliers are plotted as points.
