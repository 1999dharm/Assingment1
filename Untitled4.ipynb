14. How is memory managed in Python? 
Ans. Memory management in Python is handled by the Python Memory Manager, which is part of the Python runtime system. The memory manager is responsible for allocating and freeing memory as needed by the Python objects.
Python uses a reference-counting system to keep track of the references to objects in memory. When an object is no longer being referenced by any part of the program, its reference count drops to zero and the memory manager automatically frees the memory associated with that object.
Additionally, Python uses a garbage collector to collect and remove objects that are no longer being used by the program, even if their reference count is not zero. The garbage collector runs periodically and frees up memory that is no longer needed by the program.
In summary, Python's memory management system is automatic and efficient, freeing the programmer from having to manually manage memory allocation and freeing.


15. How to install Python on Windows and set path variables?
Ans. Download the latest version of Python from the official website (https://www.python.org/downloads/).
1. Run the installation file and follow the on-screen instructions. Make sure to select the option to add Python to your system PATH.
2. Open the Command Prompt and type python to check if Python is installed and the PATH variables are set correctly.
3. To set the PATH variables, open the Start menu, right-click on Computer, and select Properties.
4. Click on Advanced system settings and then click on the Environment Variables button.
5. In the System Variables section, scroll down and find the Path variable, then click on Edit.
6. Add the path to your Python installation (e.g. C:\Python3x\) to the Path variable.
7. Restart the Command Prompt for the changes to take effect.
