Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test.

Ans A t-test and a z-test are both statistical tests used to determine whether there is a significant difference between two means. However, they differ in the assumptions they make about the population and sample data.

A z-test is used when the population standard deviation is known, and the sample size is large (typically greater than 30). In contrast, a t-test is used when the population standard deviation is unknown and must be estimated from the sample data. T-tests are typically used with smaller sample sizes (less than 30).

An example scenario where a z-test might be used is if a researcher wants to determine if the mean height of a population is significantly different from a known standard height. In this case, the population standard deviation is already known, and the sample size is likely to be large, so a z-test would be appropriate.

On the other hand, an example scenario where a t-test might be used is if a researcher wants to determine if there is a significant difference in the mean test scores of two different groups of students. In this case, the population standard deviation is not known, and the sample size is likely to be smaller, so a t-test would be appropriate.





Q2: Differentiate between one-tailed and two-tailed tests.

Ans One-tailed and two-tailed tests are terms used to describe the directionality of a statistical hypothesis test.

A one-tailed test is a statistical test in which the hypothesis is directional, and the region of rejection is located entirely in one tail of the distribution. It is used when there is a specific prediction about the direction of the difference between two groups or conditions. For example, a one-tailed test might be used to determine if a new drug has a positive effect on a particular condition, where the hypothesis is that the drug will improve the condition, and the region of rejection is in the upper tail of the distribution.

A two-tailed test, on the other hand, is a statistical test in which the hypothesis is non-directional, and the region of rejection is split between both tails of the distribution. It is used when there is no specific prediction about the direction of the difference between two groups or conditions. For example, a two-tailed test might be used to determine if there is a significant difference in the height of men and women, where the hypothesis is that there is a difference in height, but not necessarily in any particular direction.

In summary, one-tailed tests are used when there is a specific directional hypothesis, while two-tailed tests are used when there is no specific directional hypothesis.





Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.

Ans Type 1 and Type 2 errors are two types of errors that can occur in hypothesis testing.

Type 1 error, also known as a false positive, occurs when the null hypothesis is rejected when it is actually true. This means that the test incorrectly concludes that there is a significant difference between the two groups being compared when there is no actual difference. The probability of making a Type 1 error is denoted by alpha (α), and it is typically set at a significance level of 0.05 or 0.01.

An example of a Type 1 error is when a medical test produces a false positive result. For instance, a patient is diagnosed with a disease even though they do not have it. This can lead to unnecessary medical treatment, causing physical, emotional, and financial harm to the patient.

Type 2 error, also known as a false negative, occurs when the null hypothesis is not rejected when it is actually false. This means that the test incorrectly concludes that there is no significant difference between the two groups being compared when there is an actual difference. The probability of making a Type 2 error is denoted by beta (β), and it depends on factors such as the sample size, effect size, and significance level.

An example of a Type 2 error is when a medical test produces a false negative result. For instance, a patient is diagnosed as healthy even though they have the disease. This can lead to a delay in diagnosis and treatment, which can be life-threatening in some cases.

In summary, a Type 1 error is a false positive, where the null hypothesis is rejected when it is actually true. A Type 2 error is a false negative, where the null hypothesis is not rejected when it is actually false. Both types of errors have implications for the validity and reliability of hypothesis testing, and researchers must carefully consider the risk of each type of error when designing and interpreting their studies.

Ans Bayes's theorem is a statistical theorem that describes the probability of an event occurring based on prior knowledge of conditions that might be related to the event. Bayes's theorem is often used to update the probability of a hypothesis based on new evidence or data.

The formula for Bayes's theorem is:

P(A|B) = P(B|A) * P(A) / P(B)

Where:

P(A|B) is the posterior probability of A given B, which is the updated probability of A after taking into account the new evidence B.
P(B|A) is the conditional probability of B given A, which is the probability of observing the evidence B if A is true.
P(A) is the prior probability of A, which is the probability of A before taking into account the new evidence B.
P(B) is the marginal probability of B, which is the probability of observing the evidence B.
Here is an example to illustrate Bayes's theorem:

Suppose a hospital has a new test for a rare disease, and the test is known to have a false positive rate of 5% and a false negative rate of 1%. The prevalence of the disease in the population is 1 in 1000. A patient tests positive for the disease. What is the probability that the patient actually has the disease?

Using Bayes's theorem, we can calculate the posterior probability of the patient having the disease given the positive test result:

P(disease|positive) = P(positive|disease) * P(disease) / P(positive)

P(positive|disease) = 1 - 0.01 = 0.99, which is the probability of a true positive result.
P(disease) = 0.001, which is the prevalence of the disease in the population.
P(positive) = P(positive|disease) * P(disease) + P(positive|no disease) * P(no disease)
= 0.99 * 0.001 + 0.05 * 0.999 = 0.05094, which is the probability of a positive test result.
Plugging in these values, we get:

P(disease|positive) = 0.99 * 0.001 / 0.05094 = 0.0194

So the probability that the patient actually has the disease given a positive test result is about 1.94%. This means that even though the patient tested positive, there is still a 98.06% chance that the patient does not have the disease, due to the low prevalence of the disease in the population and the high false positive rate of the test.

A confidence interval is a range of values that is likely to contain the true value of a population parameter, such as the population mean or proportion, with a certain level of confidence. It provides a measure of the precision of an estimate based on a sample from the population.

The formula for calculating a confidence interval for a population mean is:

CI = X̄ ± Z * (σ / √n)

CI is the confidence interval
X̄ is the sample mean
Z is the z-score corresponding to the desired level of confidence, such as 1.96 for 95% confidence level
σ is the population standard deviation, or the sample standard deviation if the population standard deviation is unknown

CI = 180 ± 1.96 * (10 / √100) = 180 ± 1.96
